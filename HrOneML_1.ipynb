````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Veri setini yükleme
# Not: Dosyanın yerel dizininizde 'Absenteeism_at_work.csv' adıyla bulunduğunu varsayıyoruz.
try:
    # Correcting the path to use the previously defined dataset_path
    df = pd.read_csv(os.path.join(dataset_path, 'Absenteeism_at_work.csv'), sep=';') # Veri setinde noktalı virgül (;) kullanıldığını varsayalım
except FileNotFoundError:
    print("HATA: 'Absenteeism_at_work.csv' dosyası bulunamadı. Lütfen dosya yolunu kontrol edin.")
    exit()

# Kolon isimlerindeki gereksiz boşlukları temizleme (Örn: 'Work load Average/day ')
df.columns = df.columns.str.strip()


df.info()
````

````python
# Modelde kullanmayacağımız kolonları çıkaralım
df_model = df.drop(columns=['ID', 'Hit target', 'Disciplinary failure'])

print("\nModelde Kullanılacak Kolonlar:")
print(df_model.columns.tolist())
````

````python
# pd.get_dummies() ile kategorik değişkenleri dönüştürme
# drop_first=True, çoklu doğrusal bağlantıyı (multicollinearity) önlemek için ilk kategoriyi düşürür.
df_processed = pd.get_dummies(df_model, columns=[
    'Reason for absence', 'Month of absence', 'Day of the week',
    'Seasons', 'Education', 'Social drinker', 'Social smoker'
], drop_first=True)

# 'Son' ve 'Pet' değişkenleri zaten sayısal olduğu için şimdilik dönüştürülmedi (Bağımlı değişken olarak kabul edilebilir).

print("\nOne-Hot Encoding Sonrası Kolon Sayısı:")
print(len(df_processed.columns))
print("One-Hot Encoding Sonrası İlk 5 Satır:")
print(df_processed.head())
````

````python
# Y: Bağımlı değişken (Absenteeism time in hours)
Y = df_processed['Absenteeism time in hours']

# X: Bağımsız değişkenler (Y hariç tüm kolonlar)
X = df_processed.drop('Absenteeism time in hours', axis=1)

print(f"\nX Matris Boyutu: {X.shape}")
print(f"Y Vektör Boyutu: {Y.shape}")
````

````python
# Verinin %70'i eğitim, %30'u test için kullanılacak
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

print(f"\nEğitim Seti Boyutu (X_train): {X_train.shape[0]} satır")
print(f"Test Seti Boyutu (X_test): {X_test.shape[0]} satır")
````

````python
# Gerekli kütüphaneleri (tekrar) içe aktarın, eğer Notebook'u yeniden başlattıysanız
from sklearn.model_selection import train_test_split
# ... (df_processed ve X, Y değişkenlerinin tanımlandığını varsayarak)

# Y: Bağımlı değişken (Absenteeism time in hours)
Y = df_processed['Absenteeism time in hours']

# X: Bağımsız değişkenler (Y hariç tüm kolonlar)
X = df_processed.drop('Absenteeism time in hours', axis=1)

# Verinin %70'i eğitim, %30'u test için kullanılacak
# Bu satırı ÇALIŞTIRMALISINIZ
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

print("X_train, X_test, Y_train, Y_test değişkenleri başarıyla tanımlandı.")
````

````python
# Doğrusal Regresyon modelini başlatma
model = LinearRegression()

# Modeli eğitim seti üzerinde eğitme (Şimdi çalışması gerekir)
model.fit(X_train, Y_train)

print("\nDoğrusal Regresyon Modeli Başarıyla Eğitildi.")
````

````python
# Modeli test seti üzerinde çalıştırma ve tahmin yapma
Y_pred = model.predict(X_test)

# --- Performans Metriklerini Hesaplama ---

# R-kare (R-squared) skoru: Modelin varyansı açıklama yüzdesi
r2 = r2_score(Y_test, Y_pred)

# Ortalama Karesel Hata Kökü (Root Mean Squared Error - RMSE): Tahminlerin ortalama hatası
# MSE (Mean Squared Error) hesaplanır, sonra kökü alınır
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)

print("--- Model Performans Metrikleri ---")
print(f"R-Kare Skoru (R²): {r2:.4f}")
print(f"Ortalama Karesel Hata Kökü (RMSE): {rmse:.4f} Saat")
print(f"Gerçek Y_test değerlerinin ortalaması: {Y_test.mean():.4f} Saat")
````

# Lojistik Regresyon


Yeni Hedefimiz:Veriyi tam ortadan ikiye bölmek (Dengeli Veri Seti) modelin en iyi öğrenmesini sağlar. Sizin verinizde Medyan = 3.0 olduğu için bu harika bir eşik değeridir.Sınıf 0 (Düşük Risk): Devamsızlık <= 3 Saat (Çalışanların %50'si)Sınıf 1 (Yüksek Risk): Devamsızlık > 3 Saat (Çalışanların %50'si)Bu dönüşümü yapıp Lojistik Regresyon ile hangi çalışanların "Yüksek Risk" grubuna gireceğini tahmin eden kodu yazalım mı? Bu yöntemle başarı oranımızın (Accuracy) ciddi şekilde artacağını göreceksiniz.

````python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Not: Önceki adımlardaki 'df_processed' veri setini kullanmaya devam ediyoruz.
# Eğer oturum kapandıysa önceki ön işleme kodlarını tekrar çalıştırmanız gerekebilir.

# 1. Yeni Hedef Değişkeni Oluşturma (Eşik Değeri: Medyan = 3.0)
median_value = df_processed['Absenteeism time in hours'].median()
threshold = median_value

# np.where ile koşullu atama: Değer > 3 ise 1, değilse 0
df_processed['Excessive Absenteeism'] = np.where(df_processed['Absenteeism time in hours'] > threshold, 1, 0)

# 2. Eski Hedef Değişkeni Çıkarma
# Artık saate ihtiyacımız yok, sınıfa (0 veya 1) odaklanacağız.
# Ayrıca, 'Absenteeism time in hours' bilgisini modelden çıkarmazsak model hile yapar (data leakage).
data_class = df_processed.drop(['Absenteeism time in hours'], axis=1)

# Yeni dağılımı kontrol edelim (Dengeli olması başarımızı artırır)
print(f"Eşik Değeri (Medyan): {threshold} Saat")
print("\nSınıf Dağılımı (0: Düşük, 1: Yüksek Risk):")
print(data_class['Excessive Absenteeism'].value_counts())
print("\nDağılım Oranı:")
print(data_class['Excessive Absenteeism'].value_counts(normalize=True))
````

````python
from sklearn.preprocessing import StandardScaler

# 1. Scaler (Ölçekleyici) Nesnesini Oluşturma
scaler = StandardScaler()

# 2. Sadece EĞİTİM (Train) setine göre ölçekleyiciyi eğitme
# (Test setini görmemesi lazım, yoksa kopya çekmiş oluruz)
scaler.fit(X_train)

# 3. Hem Eğitim hem Test setlerini ölçekleme
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Modeli ÖLÇEKLENMİŞ veri ile tekrar eğitme
log_reg = LogisticRegression(max_iter=5000)
log_reg.fit(X_train_scaled, y_train)

print("\n✅ Lojistik Regresyon Modeli 'Ölçeklenmiş Veri' ile Başarıyla Eğitildi!")
````

````python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# 1. Tahmin Yapma (Ölçeklenmiş test verisi ile)
y_pred = log_reg.predict(X_test_scaled)

# 2. Başarı Metriklerini Hesaplama
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"--- Model Doğruluğu (Accuracy) ---")
print(f"Başarı Oranı: %{accuracy * 100:.2f}")

print("\n--- Karışıklık Matrisi (Confusion Matrix) ---")
print(conf_matrix)
print("(Sol Üst: Doğru 'Düşük Risk', Sağ Alt: Doğru 'Yüksek Risk')")

print("\n--- Detaylı Sınıflandırma Raporu ---")
print(classification_report(y_test, y_pred))
````

````python
# Katsayıları DataFrame'e aktarma
feature_importance = pd.DataFrame({
    'Feature': X_class.columns,
    'Coefficient': log_reg.coef_[0]
})

# Odds Ratio (Olasılık Oranı) Hesaplama: Katsayının üssünü (exp) alıyoruz
feature_importance['Odds_Ratio'] = np.exp(feature_importance['Coefficient'])

# En yüksek etkiye sahip olanları sıralama (Büyükten küçüğe)
feature_importance = feature_importance.sort_values(by='Odds_Ratio', ascending=False)

print("\n--- Yüksek Devamsızlık Riskini ARTIRAN En Güçlü 10 Faktör ---")
# Odds Ratio > 1 olanlar riski artırır.
print(feature_importance.head(10)[['Feature', 'Odds_Ratio']].to_string(index=False))

print("\n--- Yüksek Devamsızlık Riskini AZALTAN (Koruyucu) 5 Faktör ---")
# Listenin sonundakiler (En küçük Odds Ratio) riski düşürenlerdir.
print(feature_importance.tail(5)[['Feature', 'Odds_Ratio']].to_string(index=False))
````
