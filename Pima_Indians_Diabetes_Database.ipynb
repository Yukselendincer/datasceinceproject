```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Assuming the dataset contains a single CSV file, find it
csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]

if csv_files:
    csv_file_path = os.path.join(path, csv_files[0])
    df = pd.read_csv(csv_file_path)
    print(f"Loaded '{csv_files[0]}' into a DataFrame.")

    # Display the first 5 rows of the DataFrame
    print("First 5 rows of the DataFrame:")
    display(df.head())
else:
    print(f"No CSV files found in the dataset path: {path}")
```

```python
%%R
# 1. Veriyi internetten doğrudan çekelim
url <- "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
col_names <- c("Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
               "Insulin", "BMI", "DiabetesPedigree", "Age", "Outcome")
pima <- read.csv(url, col.names = col_names)

# 2. Veri Temizleme (0 olan hatalı değerleri temizleyelim)
# Glikoz ve BMI 0 olamaz, bunları analiz dışı bırakıyoruz
pima_clean <- pima[pima$Glucose > 0 & pima$BMI > 0, ]

# 3. Doğrusal Regresyon Modeli
model_glikoz <- lm(Glucose ~ BMI + Age, data = pima_clean)

# 4. Sonuçları Görüntüle
summary(model_glikoz)
```

```python
%%R
library(ggplot2)

ggplot(pima_clean, aes(x = BMI, y = Glucose)) +
  geom_point(alpha = 0.4, color = "darkblue") +
  geom_smooth(method = "lm", color = "red") +
  labs(title = "BMI ve Glikoz Seviyesi Arasındaki Doğrusal İlişki",
       x = "Vücut Kitle Endeksi (BMI)",
       y = "Glikoz Konsantrasyonu") +
  theme_minimal()
```

```python
%%R
par(mfrow = c(2, 2))
plot(model_glikoz)
```
