```julia
using Pkg
Pkg.add(["JuMP", "HiGHS", "OperationsResearchModels"])
```

```julia
using JuMP
using HiGHS

# 1. Modeli Tanımla
model = Model(HiGHS.Optimizer)

# 2. Değişkenleri Tanımla (x >= 0, y >= 0)
@variable(model, x >= 0)
@variable(model, y >= 0)

# 3. Kısıtları Tanımla
@constraint(model, 2x + y <= 20)
@constraint(model, x + 3y <= 30)

# 4. Amaç Fonksiyonunu Tanımla (Z = 5x + 4y maksimizasyonu)
@objective(model, Max, 5x + 4y)

# 5. Modeli Çöz
optimize!(model)

# 6. Sonuçları Görüntüle
println("Durum: ", termination_status(model))
println("x değeri: ", value(x))
println("y değeri: ", value(y))
println("Maksimum Kâr: ", objective_value(model))
```
