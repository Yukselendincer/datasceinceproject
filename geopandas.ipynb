```python
import geopandas as gpd

url = "https://raw.githubusercontent.com/datasets/geo-boundaries-world-110m/master/countries.geojson"
world = gpd.read_file(url)

print(world.head())
```

```python
import geopandas as gpd

# We will use the 'world' GeoDataFrame that was successfully loaded
# in a previous cell (gXKHRNP1SFGz), as 'naturalearth_lowres' is causing issues.
dunya = world # Use the already loaded 'world' GeoDataFrame

# 2. Veriyi filtrele (Sadece Türkiye'yi alalım)
turkiye = dunya[dunya.name == "Turkey"]

# 3. Merkeze (centroid) bir nokta koyalım
# Re-project to a suitable projected CRS (e.g., EPSG:3857 for a general-purpose projected CRS)
# before calculating the centroid to avoid the UserWarning and ensure accuracy.
turkiye_proj = turkiye.to_crs(epsg=3857)
merkez = turkiye_proj.centroid

# 4. Görselleştir
turkiye.plot()
```

```python
import osmnx as ox
import matplotlib.pyplot as plt

# Örnek: Beşiktaş, İstanbul'daki tüm parkları (leisure=park) çekelim
tags = {'leisure': 'park'}
beşiktaş_parklar = ox.features_from_place('Beşiktaş, İstanbul, Turkey', tags)

# Veriyi inceleyelim (Bu artık bir GeoDataFrame!)
print(beşiktaş_parklar.head())

# Harita üzerinde çizelim
beşiktaş_parklar.plot(figsize=(10, 10), color='green')
plt.title("Beşiktaş'taki Parklar")
plt.show()
```
