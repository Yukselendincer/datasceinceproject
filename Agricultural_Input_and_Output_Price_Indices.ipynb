```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Veriyi Yükle
file_path = 'AHA05.20260315231153.csv'
df = pd.read_csv(file_path)

# 2. Veri Temizleme
# 'VALUE' sütunundaki sayısal olmayan değerleri temizleyelim ve sayıya çevirelim
df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')
df = df.dropna(subset=['VALUE'])

# Analiz etmek istediğimiz spesifik bir ürün seçelim (Örn: 'Agricultural output price index')
product_name = 'Agricultural output price index'
product_df = df[df['Agricultural Product'] == product_name]

# 3. Regresyon İçin Hazırlık
X = product_df['Year'].values.reshape(-1, 1)  # Bağımsız değişken: Yıl
y = product_df['VALUE'].values               # Bağımlı değişken: Fiyat Endeksi

# 4. Modeli Eğit
model = LinearRegression()
model.fit(X, y)

# 5. Gelecek Tahmini (2026 Yılı İçin)
future_year = np.array([[2026]])
prediction_2026 = model.predict(future_year)

# 6. Görselleştirme
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Mevcut Veriler')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regresyon Hattı (Trend)')
plt.scatter(future_year, prediction_2026, color='green', marker='X', s=200, label=f'2026 Tahmini: {prediction_2026[0]:.2f}')

plt.title(f'Zaman Serisi Regresyon Analizi: {product_name}', fontsize=14)
plt.xlabel('Yıl')
plt.ylabel('Endeks Değeri (2020=100)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Sonuçları Yazdır
print(f"--- {product_name} Analiz Sonuçları ---")
print(f"Model Katsayısı (Eğim): {model.coef_[0]:.2f}")
print(f"R-Kare Değeri (Doğruluk): {model.score(X, y):.4f}")
print(f"2026 yılı için beklenen endeks değeri: {prediction_2026[0]:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1. Veriyi Oku ve Temizle
df = pd.read_csv('AHA05.20260315231153.csv')
df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')
df = df.dropna(subset=['VALUE'])

# 2. Analiz İçin İki Değişken Seçelim
# Bağımsız Değişken (X): Toplam Girdi Endeksi (Maliyetler)
# Bağımlı Değişken (y): Toplam Çıktı Endeksi (Satış Fiyatları)
input_label = 'Agricultural input price index'
output_label = 'Agricultural output price index'

x_data = df[df['Agricultural Product'] == input_label][['Year', 'VALUE']]
y_data = df[df['Agricultural Product'] == output_label][['Year', 'VALUE']]

# Yıllara göre verileri birleştirelim (Merge)
merged_df = pd.merge(x_data, y_data, on='Year', suffixes=('_Input', '_Output'))

# 3. Regresyon Modeli Hazırlığı
X = merged_df['VALUE_Input'].values.reshape(-1, 1)
y = merged_df['VALUE_Output'].values

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# 4. Görselleştirme
plt.figure(figsize=(10, 6))
sns.regplot(x=X, y=y, scatter_kws={'s':100, 'color':'darkblue'}, line_kws={'color':'red'})

plt.title(f'Korelasyon Analizi: Girdi vs Çıktı Fiyatları', fontsize=14)
plt.xlabel('Girdi Fiyat Endeksi (Maliyet)', fontsize=12)
plt.ylabel('Çıktı Fiyat Endeksi (Satış)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 5. İstatistiksel Sonuçlar
r2 = r2_score(y, y_pred)
print(f"--- Regresyon Sonuçları ---")
print(f"R-Kare (Açıklayıcılık Oranı): %{r2*100:.2f}")
print(f"Eğim (Coefficient): {model.coef_[0]:.4f}")
print(f"Yorum: Girdi maliyetlerindeki her 1 birimlik artış, çıktı fiyatlarını yaklaşık {model.coef_[0]:.2f} birim etkiliyor.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Veri Hazırlığı
df = pd.read_csv('AHA05.20260315231153.csv')
df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')
df = df.dropna(subset=['VALUE'])

# Analiz edilecek kalemleri tablodaki isimlere göre seçelim
# Örn: Girdi (Yem vb.) ve Çıktı (Hayvansal Ürünler)
girdi_kalemi = 'Compound feeding stuffs'
cikti_kalemi = 'Animal output'

x_vals = df[df['Agricultural Product'] == girdi_kalemi].sort_values('Year')['VALUE'].values
y_vals = df[df['Agricultural Product'] == cikti_kalemi].sort_values('Year')['VALUE'].values

# 2. İstatistiksel Regresyon Hesaplama
slope, intercept, r_value, p_value, std_err = stats.linregress(x_vals, y_vals)

# 3. Görselleştirme
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, 'o', label='Gerçek Veriler', markersize=10, color='royalblue')
plt.plot(x_vals, intercept + slope*x_vals, 'r', label='Regresyon Çizgisi', linewidth=2)

plt.title(f'Regresyon Analizi: {girdi_kalemi} vs {cikti_kalemi}', fontsize=12)
plt.xlabel(f'{girdi_kalemi} Fiyat Endeksi', fontsize=10)
plt.ylabel(f'{cikti_kalemi} Fiyat Endeksi', fontsize=10)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 4. Raporlama
print(f"--- Analiz Raporu ---")
print(f"Korelasyon Katsayısı (R): {r_value:.4f}")
print(f"Belirleyicilik Katsayısı (R²): {r_value**2:.4f}")
print(f"P-Değeri: {p_value:.5f}")

if p_value < 0.05:
    print("Sonuç: İstatistiksel olarak anlamlı bir ilişki var.")
else:
    print("Sonuç: İlişki istatistiksel olarak zayıf veya anlamlı değil.")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# 1. Veriyi Hazırlama ve Temizleme
df = pd.read_csv('AHA05.20260315231153.csv')
df['VALUE'] = pd.to_numeric(df['VALUE'], errors='coerce')
df = df.dropna(subset=['VALUE'])

# Yardımcı fonksiyon: Belirli bir ürünün verilerini getirir
def get_val(label):
    return df[df['Agricultural Product'] == label].sort_values('Year')[['Year', 'VALUE']]

# Doğru etiketleri kullanalım
out = get_val('Agricultural output price index')
feed = get_val('Feeding stuffs')
energy = get_val('Energy') # Dosyadaki tam adı sadece 'Energy'

# 2. Verileri Yıllara Göre Birleştirme
m_df = out.merge(feed, on='Year', suffixes=('_Out', '_Feed'))
m_df = m_df.merge(energy, on='Year')
m_df.rename(columns={'VALUE': 'VALUE_Energy'}, inplace=True)

# Verinin boş olup olmadığını kontrol edelim
if m_df.empty:
    print("HATA: Eşleşen veri bulunamadı. Lütfen ürün isimlerini kontrol edin.")
else:
    # 3. Çoklu Regresyon Kurulumu
    # X: Bağımsız Değişkenler (Yem ve Enerji)
    # y: Bağımlı Değişken (Çıktı Fiyatı)
    X = m_df[['VALUE_Feed', 'VALUE_Energy']]
    y = m_df['VALUE_Out']

    multi_model = LinearRegression()
    multi_model.fit(X, y)

    # 4. Sonuçları Yazdır
    print(f"--- Çoklu Regresyon Sonuçları ---")
    print(f"Yeni R² Değeri: {multi_model.score(X, y):.4f}")
    print(f"Yem Fiyatı Katsayısı: {multi_model.coef_[0]:.4f}")
    print(f"Enerji Fiyatı Katsayısı: {multi_model.coef_[1]:.4f}")
    print("\nYorum: Bu model, çıktı fiyatlarındaki değişimi %" + str(round(multi_model.score(X, y)*100, 2)) + " oranında açıklıyor.")
```
